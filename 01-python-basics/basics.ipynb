```python
def greet(name):
    return f"Hello {name}"

print(greet("Jaspal"))
```

```python
class Person:
    def __init__(self, name):
        self.name = name
    def greet (self):
        return f"Hello my name is {self.name}"
```

```python
p = Person("jaspal")
```

```python
p.greet()
```

```python
Person("Tom").greet()
```
