```python
def greet(name):
    return f"Hello {name}"

print(greet("Jaspal"))
```

```python
class Person:
    def __init__(self, name):
        self.name = name
    def greet (self):
        return f"Hello my name is {self.name}"
```

```python
p = Person("jaspal")
```

```python
p.greet()
```

```python
Person("Tom").greet()
```

```python
numbers = [1, 2, 3, 4, 5]

squares = [n * n for n in numbers]

print(squares)
```

```python
even_numbers = [n for n in numbers if n % 2 == 0]

print(even_numbers)
```

```python
people = [
    Person("Alice"),
    Person("Bob"),
    Person("Charlie")
]

names = [p.name for p in people]

print(names)
```
