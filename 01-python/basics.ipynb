```python
def greet(name):
    return f"Hello {name}"

print(greet("Jaspal"))
```

```python
def greet(name, age = None):
    if age:
        return f"Hello my name is {name} and age is {age}"
    return f"Hello {name}"

print(greet("Jaspal"))
print(greet("Jaspal",20))
```

```python
class Person:
    def __init__(self, name):
        self.name = name
    def greet (self):
        return f"Hello my name is {self.name}"
```

```python
p = Person("jaspal")
```

```python
p.greet()
```

```python
Person("Tom").greet()
```

```python
numbers = [1, 2, 3, 4, 5]

squares = [n * n for n in numbers]

print(squares)
```

```python
even_numbers = [n for n in numbers if n % 2 == 0]

print(even_numbers)
```

Error handling - Try/catch block - Similar to Java, python also has try catch block for catching checked and unchecked exceptions

```python
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"


print(divide(10, 2))
print(divide(10, 0))
```

File handling - How to read/write from files in Python

```python
# Write to file
with open("sample.txt", "w") as f:
    f.write("Hello from Python\n")

# Read from file
with open("sample.txt", "r") as f:
    content = f.read()

print(content)
```

We can also manage file with JSON type of data

```python
import json
user = {
    "name": "Alice",
    "age": 30,
    "active": True
}

# Write JSON
with open("user.json", "w") as f:
    json.dump(user, f)

# Read JSON
with open("user.json", "r") as f:
    loaded_user = json.load(f)

print(loaded_user)
```
